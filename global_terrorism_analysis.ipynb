```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('global_terrorism.csv')
df.head()
```

```python
df.columns.shape[0]
```

```python
df=df[['iyear','imonth','iday','country_txt','region_txt','provstate','city','latitude','longitude','location','summary','attacktype1_txt','targtype1_txt','gname','motive','weaptype1_txt','nkill','nwound','addnotes']]
```

```python
df.head()
```

```python
df.rename(columns={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country",
                   "region_txt":"Region","provstate":"Province/State","city":"City",
                   "latitude":"Latitude","longitude":"Longitude","location":"Location",
                   "summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type",
                   "gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type",
                   "nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Killed']=df['Killed'].fillna(0)
df['Wounded']=df['Wounded'].fillna(0)
```

```python
df['Killed']=df['Killed'].astype(int)
df['Wounded']=df['Wounded'].astype(int)
```

```python
df.isnull().sum()
```

```python
df['Casualty']=df['Killed']+df['Wounded']
```

```python
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
# Number of attacks - yearwise
attack_yearwise = df['Year'].value_counts().sort_index()
attack_yearwise
```

```python
# Number of attacks - yearwise (Last 5)
attack_yearwise.head()
```

```python
# Number of attacks - yearwise (Recent 5)
attack_yearwise.tail()
```

```python
attack_yearwise.plot(kind='bar',figsize=(14,6), fontsize = 13)
plt.title('Yearwiwse Global Terrorist Attacks', fontsize = 17)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('No of Attacks', fontsize = 15)
plt.show()
```

```python
# Number of casualties - yearwise
casualty_yearwise = df[['Year','Casualty']].groupby('Year').sum()
casualty_yearwise
```

```python
# Number of casualties - yearwise (Last 5)
casualty_yearwise.head()
```

```python
# Number of casualties - yearwise (Recent 5)
casualty_yearwise.tail()
```

```python
casualty_yearwise.plot(kind='bar',figsize=(14,6),fontsize=13)
plt.title('Yearwise Global Casualties due to Terrorism', fontsize = 17)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('No of Casualties', fontsize = 15)
plt.show()
```

```python
# Number of people killed - yearwise (Last 5)
killed_yearwise = df[['Year','Killed']].groupby('Year').sum()
killed_yearwise.head()
```

```python
# Number of people wounded - yearwise (Last 5)
wounded_yearwise = df[['Year','Wounded']].groupby('Year').sum()
wounded_yearwise.head()
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(2,1,1)
ax1 = fig.add_subplot(2,1,2)

killed_yearwise.plot(kind="bar", figsize=(14,14), fontsize=13, ax=ax0)
ax0.set_title("People Killed in each Year due to Terrorism", fontsize=17)
ax0.set_xlabel("Year", fontsize=15)
ax0.set_ylabel("No of People Killed", fontsize=15) 

wounded_yearwise.plot(kind="bar", figsize=(14,14), fontsize=13, ax=ax1)
ax1.set_title("People Wounded in each Year due to Terrorism", fontsize=17)
ax1.set_xlabel("Year", fontsize=15)
ax1.set_ylabel("No of People Wounded", fontsize=15)

plt.subplots_adjust(hspace=0.4)

plt.show()
```
